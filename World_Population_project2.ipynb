```python
# Calculate percentage growth from 2000 to 2022
df['Population Growth 2000-2022'] = ((df['2022 Population'] - df['2000 Population']) / df['2000 Population']) * 100

print(df[['Country/Territory', 'Population Growth 2000-2022']].head())
```

```python
# Calculate percentage growth from 1070 to 2000
df['Population Growth 1970-2000'] = ((df['2000 Population'] - df['1970 Population']) / df['1970 Population']) * 100

print(df[['Country/Territory', 'Population Growth 1970-2000']].head())
```

```python
# Calculate percentage growth from 1970 to 2022
df['Population Growth 1970-2022'] = ((df['2022 Population'] - df['1970 Population']) / df['1970 Population']) * 100

print(df[['Country/Territory', 'Population Growth 1970-2022']].head())
```

```python
# Normalize Density per km² using area
df['Density Ratio'] = df['Density (per km²)'] / df['Area (km²)']

print(df[['Density Ratio']].head())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Assuming `df` is your DataFrame containing the features.
features = ['Rank', '2022 Population', '2020 Population', '2015 Population',
       '2010 Population', '2000 Population', '1990 Population',
       '1980 Population', '1970 Population', 'Area (km²)', 'Density (per km²)',
       'Growth Rate', 'World Population Percentage',
       'Population Growth 2000-2022', 'Population Growth 1970-2000',
       'Population Growth 1970-2022', 'Density Ratio', 'Continent_Asia',
       'Continent_Europe', 'Continent_North America', 'Continent_Oceania',
       'Continent_South America', 'Country_Code', 'CCA3_code', 'Capital_code']

df = df[features]

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Identify highly correlated features (threshold > 0.9)
threshold = 0.9
high_corr_pairs = []
for i in range(len(correlation_matrix.columns)):
    for j in range(i):
        if abs(correlation_matrix.iloc[i, j]) > threshold:
            col_name_1 = correlation_matrix.columns[i]
            col_name_2 = correlation_matrix.columns[j]
            high_corr_pairs.append((col_name_1, col_name_2))

print("Highly correlated features (correlation > 0.9):")
print(high_corr_pairs)

# Use VIF to identify multicollinearity
X = df.select_dtypes(include=[np.number])
vif_data = pd.DataFrame()
vif_data["Feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print("Variance Inflation Factor (VIF):")
print(vif_data)

# Drop one feature from each pair of highly correlated features
redundant_features = [pair[1] for pair in high_corr_pairs]
df_reduced = df.drop(columns=redundant_features)

print("Features after removing redundancy:")
print(df_reduced.columns)
```

```python
# Automatically identify boolean columns
bool_columns = df.select_dtypes(include=['bool']).columns

# Convert all boolean features to numerical (0/1)
df[bool_columns] = df[bool_columns].astype(int)

print("Boolean features converted to numerical:")
print(df.head())
```

Feature selection

```python
# Define the target feature (Future Population)
target_feature = '2022 Population'  # Replace with other years if needed

# Define the selected features (Excluding the target and irrelevant columns)
selected_features = [
    '2020 Population', '2015 Population', '2000 Population',  # Historical population data
    'Growth Rate', 'Population Growth 2000-2022', 'Population Growth 1970-2022',  # Growth-related features
    'Area (km²)', 'Density (per km²)', 'Density Ratio',  # Area and density-related features
    'World Population Percentage',  # Global comparison
    'Continent_Asia', 'Continent_Europe', 'Continent_North America'  # Geographical features (you can adjust depending on focus)
]

# Make sure to exclude the target feature from the selected features
selected_features = [feature for feature in selected_features if feature != target_feature]

# Extract the feature set (X) and the target variable (y)
X = df[selected_features]  # Feature set
y = df[target_feature]  # Target feature (2022 Population or another year)

# Now you can use X and y for training your model
print("Selected Features:")
print(X.columns)
print("Target Feature:")
print(target_feature)
```

```python
# Splitting the dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Feature Scaling
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Model Building
# Train a Linear Regression model
model = LinearRegression()
model.fit(X_train_scaled, y_train)

# Predict on the test set
y_pred = model.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python

# Calculate evaluation metrics
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
```

```python
# Display the results
print(f"Model Evaluation Metrics:")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
print(f"R-squared (R²): {r2:.4f}")
```

```python
# Plotting actual vs predicted values
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue', alpha=0.6)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--')  # Ideal line
plt.title('Actual vs Predicted: Population (2022)')
plt.xlabel('Actual Population')
plt.ylabel('Predicted Population')
plt.show()
```
